## Month-Wise Finding the popular products 

```python
# To find the popular products per month first we filter the data as monthly 
data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate']).dt.month_name()
data
```

```python
filtered=data[data['Description'].notnull()]
filtered
```

```python
# Removing unnecessary columns
del(filtered['InvoiceNo'])
filtered
```

```python
filtered =filtered.rename(columns={'InvoiceDate': 'Month'})
filtered
```

```python
# Grouping data monthly
Monthly = filtered.groupby(['Month', 'StockCode'])
Monthly.head()
```

```python
# Calculate total quantity sold per product for each month
sold_unit = Monthly['Quantity'].sum().unstack()
sold_unit
```

```python
# Sort products within each month by total quantity sold (descending)
sold_unit = sold_unit.sort_values(by=list(sold_unit), ascending=False)
sold_unit
```

```python
# Printing top monthly products
print(sold_unit.head(30))
```

## Defining Function

```python
def MPopular(month,n): # Defining parameters of function
    Monthly_Popular=filtered[filtered['Month'] == month]  # Filtering data according to parameters
    Monthly_Popular=Monthly_Popular.sort_values(by='Quantity',ascending=False)   # Sorting data in decreasing order of Quantity
    Monthly_Popular = Monthly_Popular.drop(columns = ["CustomerID",'Country','Quantity'])   # Removing unnecessary columns
    print(Monthly_Popular.head(n))   # Returning the desired output
    
```

```python
# But, what if user want to search country and month popular data same time 
# so now we define a function that return the popular products of particular month and country
def C_M_Popular(country,month,n): # Defining parameters of functions
    popular_products=filtered[(filtered['Month'] == month) & (filtered['Country'] == country)]  # Filtering data according to parameters
    popular_products=popular_products.sort_values(by='Quantity',ascending=False)   # Sorting data in decreasing order of Quantity
    popular_products=popular_products.drop(columns = ["CustomerID",'Quantity'])    # Removing unnecessary columns
    print(popular_products.head(n))   # Returning the desired output

     
```

```python
# products popular in particular month and country
C_M_Popular(country='United Kingdom',month='December',n=100)
```
